```python
#
# PARTICIPO ALTERACION ORDEN ULTIMO A√ëO
#
df["ParticipoAlteracionOrdenUltAnoDescripcion"] = df["ParticipoAlteracionOrdenUltAnoDescripcion"].replace(
    ["SI - Alteraci√≥n del orden SIN heridos, ni rehenes",
    "SI - Alteraci√≥n del orden con da√±os"],
    "S√≠ - leve/moderado"
)

df["ParticipoAlteracionOrdenUltAnoDescripcion"] = df["ParticipoAlteracionOrdenUltAnoDescripcion"].replace(
    ["SI - Alteraci√≥n del orden con heridos o muertos",
     "SI - Alteraci√≥n del orden con rehenes"],
    "S√≠ - grave"
)
```

```python
#
# TENTATIVAS FUGAS EVASIONES
#
df["TentativaFugasEvasionesDescripcion"] = df["TentativaFugasEvasionesDescripcion"].replace(
    ["SI. tentativa de fuga", "SI. tentativa de evasi√≥n"],
    "S√≠"
)
```

```python
#
# TUVO SALIDAS TRANSITORIAS
#
df["TuvoSalidasTransitoriasDescripcion"] = df["TuvoSalidasTransitoriasDescripcion"].replace(
    ["Si - goza de salidas transitorias"],
    "S√≠ - vigente"
)

df["TuvoSalidasTransitoriasDescripcion"] = df["TuvoSalidasTransitoriasDescripcion"].replace(
    ["Suspendidas durante el √∫ltimo a√±o", "Revocadas durante el √∫ltimo a√±o"],
    "Suspendidas/Revocadas"
)

df["TuvoSalidasTransitoriasDescripcion"] = df["TuvoSalidasTransitoriasDescripcion"].replace(
    ["No le fueron otorgadas"],
    "No otorgadas"
)
```

```python
#
# INCORPORADO REGIMEN SEMI LIBERTAD
#
# S√≠ - vigente
df["IncorporadoRegSemiLibertadDescripcion"] = df["IncorporadoRegSemiLibertadDescripcion"].replace(
    ["Si est√° incorporado/a"],
    "S√≠ - vigente"
)

# Susp/Revoc
df["IncorporadoRegSemiLibertadDescripcion"] = df["IncorporadoRegSemiLibertadDescripcion"].replace(
    ["Suspendido/a durante el √∫ltimo a√±o", "Revocado durante el √∫ltimo a√±o"],
    "Suspendido/Revocado"
)

# No incorporado
df["IncorporadoRegSemiLibertadDescripcion"] = df["IncorporadoRegSemiLibertadDescripcion"].replace(
    ["No est√° incorporado/a al r√©gimen de semilibertad"],
    "No incorporado"
)
```

```python
#
# PARTICIPA PROGRAMA PRISION DISCONTINUA
#
# S√≠ - vigente
df["ParticipaProgramaPrisionDiscontinuaDescripcion"] = df["ParticipaProgramaPrisionDiscontinuaDescripcion"].replace(
    ["Si est√° incorporado/a"],
    "S√≠ - vigente"
)

# Renuncia/Revocada
df["ParticipaProgramaPrisionDiscontinuaDescripcion"] = df["ParticipaProgramaPrisionDiscontinuaDescripcion"].replace(
    ["Renuncia del interno/a durante el √∫ltimo a√±o", "Revocada durante el √∫ltimo a√±o"],
    "Renuncia/Revocada"
)

# No incorporado
df["ParticipaProgramaPrisionDiscontinuaDescripcion"] = df["ParticipaProgramaPrisionDiscontinuaDescripcion"].replace(
    ["No est√° incorporado/a"],
    "No incorporado"
)
```

```python
#
# PARTICIPA PROGRAMA SEMI DETENCION
#
# S√≠ - vigente (Diurna + Nocturna)
df["ParticipaProgramaSemiDetencionDescripcion"] = df["ParticipaProgramaSemiDetencionDescripcion"].replace(
    ["Si. Prisi√≥n Diurna", "Si. Prisi√≥n Nocturna"],
    "S√≠ - vigente"
)

# Revocada
df["ParticipaProgramaSemiDetencionDescripcion"] = df["ParticipaProgramaSemiDetencionDescripcion"].replace(
    ["Revocada durante el √∫ltimo a√±o"],
    "Revocada"
)

# No incorporado
df["ParticipaProgramaSemiDetencionDescripcion"] = df["ParticipaProgramaSemiDetencionDescripcion"].replace(
    ["No tiene semidetenci√≥n"],
    "No incorporado"
)
```

# ONE-HOT ENCODING

Para el ONE-HOT ENCODING de abajo, seguimos la siguiente estructuraca (DROP_FIRST) en caso de que tenga mas de 2 valores posible (como SEXO = varon - mujer - otro)

SexoDescripcion_Mujer    SexoDescripcion_Otro
---------------------    --------------------
        0                        0           ‚Üê Esto es "Var√≥n" (impl√≠cito)
        1                        0           ‚Üê Esto es "Mujer"
        0                        1           ‚Üê Esto es "Otro



Se crea N-1 columnas de la cantidad de categorias de una variable:
Nacionalidad: 8 categor√≠as ‚Üí 7 columnas
Estado Civil: 6 categor√≠as ‚Üí 5 columnas

```python
# ============================================================
# ONE-HOT ENCODING
# ============================================================

# Columnas num√©ricas que NO hay que encodear
numeric_cols = ['Edad', 'ParticipacionProgramaLaboral', 'ParticipacionActividadesDeportivas',
                'RecibioVisitasUltimoAno', 'TipoCondena', 'ParticipaProgramaPreLibertad']

# Todas las dem√°s son categ√≥ricas
cat_cols = [col for col in df.columns if col not in numeric_cols]

print("=" * 60)
print("APLICANDO ONE-HOT ENCODING")
print("=" * 60)
print(f"Columnas categ√≥ricas a encodear: {len(cat_cols)}")
print(f"Columnas num√©ricas (se mantienen): {len(numeric_cols)}")

# Aplicar one-hot encoding
df_encoded = pd.get_dummies(
    df,
    columns=cat_cols,
    drop_first=True,  # Evita multicolinealidad (elimina 1 categor√≠a de cada variable)
    dtype=int
)

print("\n" + "=" * 60)
print("RESULTADO FINAL")
print("=" * 60)
print(f"Filas: {len(df_encoded)}")
print(f"Columnas ANTES: {len(df.columns)}")
print(f"Columnas DESPU√âS: {len(df_encoded.columns)}")
print(f"Nuevas columnas creadas: {len(df_encoded.columns) - len(df.columns)}")

# Ver primeras columnas
print(f"\nPrimeras 10 columnas:")
print(df_encoded.columns[:10].tolist())
```

# ANALISIS COLUMNA REINCIDENTES

```python
# Buscar la columna de reincidencia en df_encoded
reincidencia_cols = [col for col in df_encoded.columns if 'Reincidente' in col]
print("Columnas de Reincidencia encontradas:")
print(reincidencia_cols)

# Ver distribuci√≥n de la variable objetivo
for col in reincidencia_cols:
    print(f"\n{col}:")
    print(df_encoded[col].value_counts())
    print(f"Nulos: {df_encoded[col].isnull().sum()}")
```

Hay 2 columnas originales de reincidencia, por lo que son columnas criticas que son:
*   EsReincidenteDescripcion_Reincidente
*   EsReincidenteDescripcion_Reiterante

Lo que hicimos es unificar este criterio, es decir, que si cualquiera de ambas columnas=1, la suma como una probabilidad de reincidir.

```python
# Combinamos columnas
df_encoded['es_reincidente'] = (
    (df_encoded['EsReincidenteDescripcion_Reincidente (art. 50 CP)'] == 1) |
    (df_encoded['EsReincidenteDescripcion_Reiterante'] == 1)
).astype(int)

print("Nueva variable combinada:")
print(df_encoded['es_reincidente'].value_counts())
```

En base al fragmento de c√≥digo de arriba, dejamos explicada la diferencia entre Reincidente y Reiterante:


*   Reincidente: Persona que comete un nuevo delito despu√©s de haber sido   condenada por sentencia firme por un delito anterior. Es decir: cometi√≥ delito ‚Üí fue condenado ‚Üí cumpli√≥ o est√° cumpliendo pena ‚Üí vuelve a cometer delito.


*   Reiterante: Persona que cometi√≥ m√∫ltiples delitos antes de ser juzgado por el primero. Es decir: cometi√≥ delito 1, delito 2, delito 3 ‚Üí reci√©n ah√≠ es atrapado y juzgado por todos juntos.



Eliminamos ambas columnas porque ya estan contenidas por la tercera columna creada (ser√≠a redundancia de datos y crea un margen de error).
La Y es lo que quer√©s entender o predecir, en este caso la reincidencia. Es tu variable objetivo, la respuesta a la pregunta: ¬øEsta persona va a reincidir?

La X son todas las caracter√≠sticas que podr√≠an explicar esa reincidencia: edad, tipo de delito, nivel educativo, situaci√≥n laboral, conducta en prisi√≥n, etc. Son las variables explicativas o features.

```python
# Eliminar las columnas originales de reincidencia (ya tenemos la combinada)
columnas_reincidencia_original = ['EsReincidenteDescripcion_Reincidente (art. 50 CP)',
                                    'EsReincidenteDescripcion_Reiterante']

# Separar X (features) e y (target)
y = df_encoded['es_reincidente']
X = df_encoded.drop(columns=columnas_reincidencia_original + ['es_reincidente'])

print("=" * 60)
print("Separacion de variables objetivas ")
print("=" * 60)
print(f"Features (X): {X.shape[0]} filas x {X.shape[1]} columnas")
print(f"Target (y): {y.shape[0]} valores")
print(f"\nDistribuci√≥n del target:")
print(y.value_counts(normalize=True))
print(f"\nPrimeras 5 columnas de X:")
print(X.columns[:5].tolist())
```

# ANALISIS DE ASOCIACION

El an√°lisis de asociaci√≥n o bivariado busca identificar qu√© caracter√≠sticas individuales est√°n relacionadas con la reincidencia. Para cada variable se compara la tasa de reincidencia entre quienes tienen esa caracter√≠stica versus quienes no la tienen.

```python
# ================================================================
# AN√ÅLISIS DE ASOCIACI√ìN CON REINCIDENCIA
# ================================================================

import numpy as np
import pandas as pd

# Funci√≥n para calcular tasas de reincidencia por variable
def analizar_asociacion(X, y):
    resultados = []

    for col in X.columns:
        if X[col].nunique() == 2:  # Variables binarias (0/1)
            tasa_cuando_1 = y[X[col] == 1].mean() * 100
            tasa_cuando_0 = y[X[col] == 0].mean() * 100
            diferencia = tasa_cuando_1 - tasa_cuando_0
            n_casos_1 = (X[col] == 1).sum()

            resultados.append({
                'Variable': col,
                'Tasa_SI': round(tasa_cuando_1, 2),
                'Tasa_NO': round(tasa_cuando_0, 2),
                'Diferencia': round(diferencia, 2),
                'N_casos': n_casos_1
            })

    return pd.DataFrame(resultados)

# Ejecutar an√°lisis
print("=" * 80)
print("AN√ÅLISIS: FACTORES QUE M√ÅS AUMENTAN O DISMINUYEN LA REINCIDENCIA")
print("=" * 80)

resultados = analizar_asociacion(X, y)

# Ordenamientos separados
aumentan = resultados.sort_values('Diferencia', ascending=False).head(15)
disminuyen = resultados.sort_values('Diferencia', ascending=True).head(15)

print("\nTOP 15 FACTORES QUE M√ÅS AUMENTAN LA REINCIDENCIA:")
print("=" * 80)
print(aumentan.to_string(index=False))

print("\n\nTOP 15 FACTORES QUE M√ÅS DISMINUYEN LA REINCIDENCIA:")
print("=" * 80)
print(disminuyen.to_string(index=False))

print("\n\nTASA GENERAL DE REINCIDENCIA EN EL DATASET:")
print(f"{y.mean() * 100:.2f}%")
```

```python
# ================================================================
# VISUALIZACI√ìN: TOP FACTORES DE RIESGO (AUMENTAN REINCIDENCIA)
# ================================================================

import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionamos los 10 factores con mayor impacto positivo en la reincidencia
top_10_aumento = aumentan.head(10)

# Grafico 1: Diferencia en % de reincidencia
plt.figure(figsize=(14, 8))
plt.barh(range(len(top_10_aumento)), top_10_aumento['Diferencia'], color='crimson')
plt.yticks(range(len(top_10_aumento)),
           [v[:50] + '...' if len(v) > 50 else v for v in top_10_aumento['Variable']],
           fontsize=10)
plt.xlabel('Diferencia en % de reincidencia', fontsize=12)
plt.title('TOP 10 FACTORES QUE M√ÅS AUMENTAN LA REINCIDENCIA', fontsize=14, fontweight='bold')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

# Grafico 2: Comparaci√≥n Tasa_SI vs Tasa_NO
fig, ax = plt.subplots(figsize=(14, 8))
x = range(len(top_10_aumento))
width = 0.35

ax.barh([i - width/2 for i in x], top_10_aumento['Tasa_SI'], width, label='Con factor', color='crimson')
ax.barh([i + width/2 for i in x], top_10_aumento['Tasa_NO'], width, label='Sin factor', color='lightgreen')

ax.set_yticks(x)
ax.set_yticklabels([v[:50] + '...' if len(v) > 50 else v for v in top_10_aumento['Variable']], fontsize=10)
ax.set_xlabel('Tasa de reincidencia (%)', fontsize=12)
ax.set_title('COMPARACI√ìN: Reincidencia CON vs SIN el factor', fontsize=14, fontweight='bold')
ax.legend()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

# ANALISIS ETARIO

Se segment√≥ la poblaci√≥n en cinco grupos etarios para identificar si la edad est√° asociada con diferentes tasas de reincidencia. Los grupos de 26-35 a√±os y 36-45 a√±os mostraron las tasas m√°s altas (17.81% y 18.19% respectivamente), mientras que los grupos m√°s j√≥venes (18-25) y mayores (56+) presentaron tasas significativamente menores (10.66% y 9.52%). Esto sugiere que el riesgo de reincidencia alcanza su pico en la adultez temprana y media, y disminuye con la edad avanzada.

```python
# ================================================================
# AN√ÅLISIS DE EDAD Y REINCIDENCIA
# ================================================================

# Crear grupos etarios
X['grupo_edad'] = pd.cut(X['Edad'],
                         bins=[0, 25, 35, 45, 55, 100],
                         labels=['18-25', '26-35', '36-45', '46-55', '56+'])

# Calcular tasa de reincidencia por grupo etario
tasa_por_edad = y.groupby(X['grupo_edad']).agg(['mean', 'count']) * [100, 1]

print("=" * 60)
print("TASA DE REINCIDENCIA POR GRUPO ETARIO")
print("=" * 60)
print(f"{'Grupo':<12} {'Tasa %':<12} {'Cantidad':<12}")
print("-" * 60)
for grupo in tasa_por_edad.index:
    tasa = tasa_por_edad.loc[grupo, 'mean']
    cantidad = int(tasa_por_edad.loc[grupo, 'count'])
    print(f"{grupo:<12} {tasa:>6.2f}%      {cantidad:>8,}")

# Visualizaci√≥n
plt.figure(figsize=(10, 6))
tasa_por_edad['mean'].plot(kind='bar', color='steelblue')
plt.axhline(y=y.mean()*100, color='red', linestyle='--', label=f'Promedio general: {y.mean()*100:.2f}%')
plt.title('TASA DE REINCIDENCIA POR GRUPO ETARIO', fontsize=14, fontweight='bold')
plt.xlabel('Grupo etario', fontsize=12)
plt.ylabel('Tasa de reincidencia (%)', fontsize=12)
plt.xticks(rotation=0)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

# ANALISIS DE DELITOS

Los delitos contra la propiedad mostraron la tasa m√°s alta de reincidencia (22.39%), seguidos por delitos contra la libertad (21.53%). En contraste, los delitos sexuales presentaron la tasa m√°s baja (6.26%), muy por debajo del promedio general. Los delitos vinculados a drogas tuvieron una tasa de 18.56%, ligeramente superior al promedio, mientras que delitos contra las personas alcanzaron 14.67%.

```python
# ================================================================
# AN√ÅLISIS DE DELITOS Y REINCIDENCIA
# ================================================================

# Obtener columnas de delitos
delito_cols = [col for col in X.columns if 'Delito' in col and 'Descripcion' in col]

# Analizar cada tipo de delito (agrupado)
delitos_unicos = []
for col in delito_cols:
    for categoria in X[col].unique():
        if categoria == 1:  # Es una columna binaria despu√©s del encoding
            nombre_delito = col.replace('Delito1Descripcion_', '').replace('Delito2Descripcion_', '').replace('Delito3Descripcion_', '').replace('Delito4Descripcion_', '').replace('Delito5Descripcion_', '')
            delitos_unicos.append(nombre_delito)

delitos_unicos = list(set(delitos_unicos))

# Calcular tasa por categor√≠a de delito principal (Delito1)
delito1_cols = [col for col in X.columns if 'Delito1Descripcion_' in col]

print("=" * 70)
print("TASA DE REINCIDENCIA POR TIPO DE DELITO PRINCIPAL")
print("=" * 70)

tasas_delitos = {}
for col in delito1_cols:
    if X[col].sum() > 100:  # Solo categor√≠as con >100 casos
        tasa = y[X[col] == 1].mean() * 100
        cantidad = X[col].sum()
        nombre = col.replace('Delito1Descripcion_', '')
        tasas_delitos[nombre] = {'tasa': tasa, 'cantidad': cantidad}

# Ordenar por tasa
tasas_delitos_df = pd.DataFrame(tasas_delitos).T.sort_values('tasa', ascending=False)

print(f"{'Tipo de delito':<45} {'Tasa %':>10} {'N casos':>10}")
print("-" * 70)
for delito, row in tasas_delitos_df.iterrows():
    print(f"{delito[:44]:<45} {row['tasa']:>9.2f}% {int(row['cantidad']):>10,}")

# Visualizaci√≥n
plt.figure(figsize=(12, 8))
plt.barh(range(len(tasas_delitos_df)), tasas_delitos_df['tasa'], color='coral')
plt.yticks(range(len(tasas_delitos_df)), tasas_delitos_df.index, fontsize=10)
plt.xlabel('Tasa de reincidencia (%)', fontsize=12)
plt.title('TASA DE REINCIDENCIA POR TIPO DE DELITO PRINCIPAL', fontsize=14, fontweight='bold')
plt.axvline(x=y.mean()*100, color='red', linestyle='--', linewidth=2, label=f'Promedio: {y.mean()*100:.2f}%')
plt.legend()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```
